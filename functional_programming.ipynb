## <font color = 'royalblue'> Functional Programming

```python
def factorial(n: int):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(4)
```

```python
def permutation(n: int, r: int):
    if r <= 0 or r >= n:
        return 0
    else:
        return factorial(n) / factorial(n - r)

permutation(5, 2)
```

```python
def combination(n: int, r: int):
    return permutation(n, r) * (1 / factorial(r))

combination(5, 2)
```

## <font color = 'royalblue'> Games

```python
from random import randint

def generate_random_number():
    return randint(1, 100)

def check_win(number: int, guess: int) -> bool:
    return number == guess

def jugar():
    number = generate_random_number()
    lives = 5

    while lives > 0:
        guess = int(input("Give me a number: "))

        if check_win(number, guess):
            print("You win")
            break
        else:
            if guess > number:
                print(f"{guess} --> Too high")
                lives -= 1
            else:
                print(f"{guess} --> Too low")
                lives -= 1
        if lives < 1:
            print("You suck! ðŸ˜›ðŸ˜›")
            print(f'The number is {number}')
```

```python
jugar()
```
